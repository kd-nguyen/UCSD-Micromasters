# Research question:
## Does preference for a movie genre correlate with the rating for that genre?

# Importing neccessary modules

```python
data_path = r'C:\Projects\UCSD-Micromasters\1-Python_for_data_science\4-Pandas\movielens\\'

%matplotlib inline
import dask.dataframe as dd

import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import random

from functools import reduce
```

# Reading data using dask (due to memory constraint)

```python
movies_df = pd.read_csv(data_path + 'movies.csv')
movies_df.head()
```

```python
ratings_df = dd.read_csv(data_path + 'ratings.csv')
ratings_df.head()
```

### Expand compound movie genres (e.g. Action|Adventure), and melt genre columns into one column (genre)

```python
movies_expanded = pd.concat([movies_df, movies_df['genres'].str.get_dummies('|')], axis='columns')
movies_long = pd.melt(movies_expanded.drop(['title', 'genres'], axis=1), 
                      id_vars='movieId', var_name='genre', value_name='count')
movies_long_trimmed = movies_long[movies_long['count'] == 1]
```

```python
movies_long_trimmed[movies_long_trimmed['movieId'] == 1]
```

```python
movies_long_trimmed_dask = dd.from_pandas(movies_long_trimmed, npartitions=6)
```

### Merge (transformed) movie df with ratings df and keep only relevant columns (userID, rating, genre, count)

```python
merged_df = dd.merge(ratings_df, movies_long_trimmed_dask)

del merged_df['movieId']
del merged_df['timestamp']
```

```python
merged_df
```

### For each user, get mean ratings for each genre, and find the percentage of ratings the user leaves on that genre compared to his/her total ratings count

```python
grouped_df = merged_df.groupby(['userId', 'genre']).agg(
    {'rating':'mean', 'count':'sum'})

aggregated_df = grouped_df.compute()
aggregated_df['percent'] = aggregated_df.groupby('userId')['count'].transform(
    lambda x: x * 100 / x.sum())
```
